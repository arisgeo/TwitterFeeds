```python
def ReTweet(input_txt):
    if input_txt.split()[0] == 'RT':
        retweet = 1
        input_txt = input_txt[3:]
    else:
        retweet = 0
        
    return input_txt, retweet
```

```python
all_features[['text', 'retweet']] = all_features.apply(
    lambda row: pd.Series(ReTweet(row['text'])), axis=1)
all_features.head()
```

```python
# Number of Characters
all_features['tot_chars'] = all_features['text'].apply(len)
```

```python
import re
import textstat

def clean_tweet(a_string):
    # find all twitter handles
    mentions = re.findall("@[\w]*", a_string)
    
    # find all hashtags
    hashtags = re.findall("#[\w]*", a_string)
    
    # check if tweet has urls
    if re.search("(?P<url>https?://[^\s]+)", a_string) is not None:
        url = 1
    else:
        url = 0
    
    # drop mentions only
    a_string = re.sub("@[\w]*", '', a_string)
    
    # Substitute multiple whitespace with single whitespace
    a_string = ' '.join(a_string.split())
        
    return a_string, len(mentions), mentions, len(hashtags), hashtags, url, textstat.sentence_count(a_string)
```

```python
# Count twitter mentions and remove them from text
all_features[['tidy_text', 'n_ment', 'mentions',
             'n_hash', 'hashtags', 'w/url', 'sentences']] = all_features.apply(
    lambda row: pd.Series(clean_tweet(row['text'])), axis=1)
all_features.head()
```

```python
from nltk.tokenize import TweetTokenizer
import spacy

# Load the en_core_web_sm model
nlp = spacy.load('en_core_web_sm')

# Get list of stopwords
stopwords = spacy.lang.en.stop_words.STOP_WORDS

def FEATURES(a_string):
    tknzr = TweetTokenizer()
    tokens = tknzr.tokenize(a_string)
    
    ###READABILITY###
    
    # remove punctuation tokens
    no_punctuation = [item for item in tokens if len(re.sub('[^a-zA-Z0-9]', '', item)) > 0]
    
    # remove websites or other words with weird format
    words = [item for item in no_punctuation if len(re.sub('[^a-zA-Z0-9#]', '', item)) == len(item)]
    
    # Compute length of each word and average
    tot_word_length = sum( [len(word) for word in words] )
    
    # Compute syllable length of each word
    syllables = [textstat.syllable_count(word, lang='en_US') for word in words]
    
    # Compute average number of syllables per word
    tot_syllabes = sum(syllables)
    
    # Count Hard Words (more than 2 syllables)
    HW = len([i for i in syllables if i > 2])
    
    content = ' '.join(words)
    # Number of alphabetical letters in words
    alpha = sum(1 for c in content if c.isalpha())
    # Number of uppercase letters in words
    cap = sum(1 for c in content if c.isupper())
    
    
    ###POS###
    # remove hashtags
    cleaned_string = ' '.join([word for word in tokens if not word.startswith('#')])

    # Create a Doc object
    doc = nlp(cleaned_string)
    
    # List stopwords in tweet
    stop_words = [token.text for token in doc if token.text in stopwords]

    # Generate named entities
    ne = [(ent.text, ent.label_) for ent in doc.ents]

    # Generate list of tokens and pos tags
    tokens = [(token.text, token.pos_) for token in doc
           if len(re.sub('[^a-zA-Z0-9]', '', token.text)) > 0
           and len(re.sub('[^a-zA-Z0-9#]', '', token.text)) == len(token.text)]
    
    pos = [token[1] for token in tokens]
    
    # POS
    nr_nouns = pos.count('NOUN')
    nr_adj = pos.count('ADJ')
    
    nr_verbs = pos.count('VERB')
    nr_adv = pos.count('ADV')
    
    nr_pron = pos.count('PRON')
    
    
    #NER - Name Entity Recognition
    # Identify the entities
    entities = [ent.text for ent in doc.ents]
    
    # Identify people
    people = [ent.text for ent in doc.ents if ent.label_ == 'PERSON']
    
    # Identify the organisations
    organisations = [ent.text for ent in doc.ents if ent.label_ == 'ORG']
    
    return (len(words), alpha, cap, tot_word_length, tot_syllabes, HW, len(stop_words),
            nr_nouns,nr_adj, nr_verbs,nr_adv, nr_pron, len(entities), len(people), len(organisations))
```

```python
all_features[['num_words', 'alpha', 'cap', 'TCW', 'TSW', 'nHW',
            'StopW', 'nouns','adj','verbs','adv','pron','Tot_entities','People','ORGs']] = all_features.apply(lambda row:
                                                              pd.Series(FEATURES(row['tidy_text'])), axis=1)
```

```python
all_features_agg = pd.DataFrame({'tot_tweets': all_features.groupby(['screen_name'])['text'].count(),
                               'retweets': all_features.groupby(['screen_name'])['retweet'].sum()})
all_features_agg['ogtweets'] = all_features_agg['tot_tweets'] - all_features_agg['retweets']
all_features_agg['OGTratio'] = all_features_agg['ogtweets']/all_features_agg['tot_tweets']*100
all_features_agg.head()
```

```python
# check in any accounts have 0 original tweets
retweets_only = all_features_agg.loc[all_features_agg['ogtweets'] == 0]
retweets_only
```

```python
all_features_agg = all_features_agg.merge(all_features.loc[all_features['retweet']==0].groupby(['screen_name']).sum().drop(
    columns = ['tweetid','retweet','n_hash']), left_index=True, right_index=True)
all_features_agg.head()
```

```python
# Change column names
col_names = ['tot_tweets', 'retweets', 'ogtweets', 'retweet_countOT', 'favourite_countOT', 'OGTratio', 'char_OT', 'ment_OT',
       'w/urlR_OT', 'sent_OT', 'words_OT', 'alp_OT', 'cap_OT', 'TCW_OT', 'TSW_OT', 'nHW_OT', 'StopW_OT',
       'nouns_OT', 'adj_OT', 'verbs_OT', 'adv_OT', 'pron_OT', 'Tot_entities_OT', 'People_OT', 'ORGs_OT']
all_features_agg.columns = col_names

all_features_agg[col_names[4:]] = all_features_agg[col_names[4:]].div(all_features_agg.ogtweets, axis=0)
all_features_agg['w/urlR_OT'] = all_features_agg['w/urlR_OT']*100
```

```python
# Capital letters ratio
all_features_agg.insert(13,'capR_OT', all_features_agg['cap_OT']/all_features_agg['alp_OT']*100)
```

```python
# Words ratios
all_features_agg[['pHW_OT', 'pStopW_OT', 'pnouns_OT', 'padj_OT', 'pverbs_OT', 'padv_OT', 'ppron_OT']] = all_features_agg[[
    'nHW_OT', 'StopW_OT', 'nouns_OT', 'adj_OT', 'verbs_OT', 'adv_OT', 'pron_OT']].div(all_features_agg.words_OT, axis=0)*100
```

```python
# The Gunning Fog Index
all_features_agg['Gun_Index'] = 0.4 * ( all_features_agg['nouns_OT']/all_features_agg['sent_OT'] + all_features_agg['pHW_OT'] )
```

```python
all_features_agg = pd.concat([all_features_agg,retweets_only],join_axes=[all_features_agg.columns])
all_features_agg.fillna(0,inplace = True)
```
