```python
def authenticate(api_key, secret_key, access_token, secret_token):

    auth = tweepy.OAuthHandler(api_key, secret_key)
    auth.set_access_token(access_token, secret_token)
    api = tweepy.API(auth, 
                     wait_on_rate_limit=True, 
                     wait_on_rate_limit_notify=True)
  
    return api


api = authenticate(tc.api_key, tc.secret_key, tc.access_token, tc.secret_token)
```

```python
def get_200_tweets(screen_name):
    #Twitter only allows access to a users most recent 3240 tweets with   this method

    #authorize twitter, initialize tweepy
    auth = tweepy.OAuthHandler(tc.api_key, tc.secret_key)
    auth.set_access_token(tc.access_token, tc.secret_token)
    api = tweepy.API(auth)

    #make initial request for most recent tweets (200 is the maximum   allowed count)
    new_tweets = api.user_timeline(screen_name = screen_name,tweet_mode='extended',count=200)

    #transform the tweepy tweets into a 2D array
    tweets = [[tweet.id_str, tweet.created_at, tweet.full_text] for tweet in new_tweets]
    
    #transform into a df
    df = pd.DataFrame(tweets,columns=['tweetid','dt','text'])
    
    return df
```
